# Manipulação de Variáveis

##### Selecionar Colunas

```python
df[['timestamp', 'open', 'close']].head()
```

##### Operações Entre Colunas

```python
df['diff'] = df['close'] - df['open']
df.head()
```

##### Excluir uma coluna

```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df
```

##### Ordenando

```python
df.sort_values('diff').head()
```

```python
df.sort_values('diff', ascending=False).head()
```

```python
df.sort_values(['close', 'volume'], ascending=False).head()
```

##### Tratamento de missings

```python
# Usando o parametro how='any', se ALGUM dos valores são missing, remova a linha ou coluna. 
df.dropna() # Nesse caso, how='any' é o valor padrão do parâmetro.
```

```python
# Usando o parametro how='all', se TODOS os valores são missing, remova a linha ou coluna.
df.dropna(how='all')
```

```python
# Substitui o valor nulo pela média
df['open'].fillna(df['open'].mean(), inplace=False)
```

```python
# Substitui o valor nulo por um valor constante
df['open'].fillna(-999, inplace=False)
```

##### Case When

```python
df['flag'] = np.where((df['open'] > df['close']), 'baixa', 'alta')
df.head()
```

##### Filtros

```python
df[df['flag'] == 'alta']
```

```python
df[(df['flag'] == 'alta') & (df['diff'] > 0)]
```

##### Contagem de valores discretos

```python
df['flag'].value_counts()
```

```python
df['flag'].value_counts(1)
```

##### Agrupamentos 

```python
df['timestamp'].agg({'min', 'max'})
```

```python
df.groupby(['flag']).mean()
```

```python
df.groupby(['flag']).agg(min_open=('open', 'min'), max_open=('open', 'max'))
```

```python
df.groupby(['flag']).agg(min_open=('open', 'min'), max_open=('open', 'max')).reset_index()
```

```python
(
    df
    .groupby(['flag'])
    .agg(min_open=('open', 'min'), 
         max_open=('open', 'max'),
         min_close=('close', 'min'),
         max_close=('close', 'max'))
    .reset_index()
)
```

# Operações entre Tabelas

```python
df_itub = pd.read_csv(f"{pasta_raiz}/itub3.csv")
```

```python
df_petr = pd.read_csv(f"{pasta_raiz}/petr4.csv")
```

```python
df_itub['stock_code'] = 'itub3'
df_petr['stock_code'] = 'petr4'
```

##### Empilhar bases

```python
pd.concat([df_itub, df_petr])
```

##### Left Join

```python
pd.merge(df_itub, df_petr, how='left', on='timestamp')
```

##### Inner Join

```python
pd.merge(df_itub, df_petr, how='inner', on='timestamp')
```
