# Feature Engineering

```python
abt = (
    df
    .sort_values('timestamp')
    .assign(date = lambda x: x['timestamp'].dt.normalize())
    .groupby('date')
    .agg(open  = ('open', 'first'), 
         close = ('close', 'last'),
         high  = ('high', 'max'),
         low   = ('low', 'min'))
    .reset_index()
    .assign(year      = lambda x: x['date'].dt.year)
    .assign(month     = lambda x: x['date'].dt.month)
    .assign(day       = lambda x: x['date'].dt.day)
    .assign(weekday   = lambda x: x['date'].dt.weekday)
    .assign(amplitude = lambda x: x['high'] - x['low'])
    .assign(retorno   = lambda x: x['close'] / x['open'] - 1)
    .assign(close_ant = lambda x: x['close'].shift())
    .assign(label     = lambda x: np.where(x['retorno'] > 0, 1, 0))
    .filter(['year', 'month', 'day', 'weekday', 'open', 'close', 'close_ant', 'high', 'low', 'amplitude', 'retorno', 'label'])
    .dropna()
)
```

```python
abt
```

```python
X = abt.drop(['label'], axis=1)
X.head()
```

```python
y = abt['label']
```

# Model Training and Evaluation

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

model = DecisionTreeClassifier()

model.fit(X, y)
```

```python
y_pred = model.predict(X)
probs = model.predict_proba(X)
```

```python
(
    abt
    .assign(predict = y_pred)
    .assign(probs   = probs[:, 1])
)
```

```python
fig, ax = plt.subplots()

plot_tree(model, ax=ax);
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y, y_pred)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y, y_pred)
cm
```

```python
df_cm = pd.DataFrame(cm, columns=['false', 'positive'], index=['false', 'positive'])
df_cm
```

```python
fig, ax = plt.subplots(figsize=(6,4))

sns.heatmap(df_cm, cmap='Blues', annot=True, annot_kws={'size': 20}, ax = ax);
```
